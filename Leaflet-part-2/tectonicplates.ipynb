```python
# Create a map centered on a global view
mymap = folium.Map(location=[0, 0], zoom_start=2, tiles='Stamen Terrain')

# Optional: Add more base maps
folium.TileLayer('OpenStreetMap').add_to(mymap)
folium.TileLayer('Stamen Toner').add_to(mymap)
folium.TileLayer('Stamen Watercolor').add_to(mymap)
folium.TileLayer('cartodb positron').add_to(mymap)
folium.TileLayer('cartodb dark_matter').add_to(mymap)
```

```python
# Define the URL for the earthquake GeoJSON data
earthquake_url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson'

# Fetch the data
response = requests.get(earthquake_url)
data = response.json()

# Extract features from the GeoJSON data
features = data['features']

# Create a list to store earthquake data
earthquake_data = []
for feature in features:
    properties = feature['properties']
    geometry = feature['geometry']
    record = {
        'id': feature['id'],
        'time': properties['time'],
        'place': properties['place'],
        'magnitude': properties['mag'],
        'longitude': geometry['coordinates'][0],
        'latitude': geometry['coordinates'][1]
    }
    earthquake_data.append(record)

# Create a DataFrame from the list
df = pd.DataFrame(earthquake_data)

# Save the DataFrame to a CSV file
csv_file_path = 'earthquake_data.csv'
df.to_csv(csv_file_path, index=False)

print(f"Data has been saved to {csv_file_path}")

# Initialize the map
mymap = folium.Map(location=[0, 0], zoom_start=2)  # Center the map and set the initial zoom level

# Create a feature group for the earthquake layer
earthquake_layer = folium.FeatureGroup(name="Earthquakes")
```

```python
# Loop through the earthquake data and plot them on the map
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color='red',
        fill=True,
        fill_opacity=0.6,
        popup=f"Magnitude: {row['magnitude']}, Location: {row['place']}"
    ).add_to(earthquake_layer)

# Add the earthquake layer to the map
earthquake_layer.add_to(mymap)

# Add layer control to toggle between layers
folium.LayerControl().add_to(mymap)

# Save the map to an HTML file
mymap.save('earthquake_map.html')

print("Map has been saved to 'earthquake_map.html'")
```
